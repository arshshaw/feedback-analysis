```python
import pandas as pd
```

```python
list_data=[]
list_tag=[]
with open("en-training.txt") as devs:
    for i in devs:
        #print(i.strip())
        d=i.strip().lower().split()
        if len(d)>1:
            del(d[0])
            #print(d)
            tag=d[-1]
            list_tag.append(tag)
            del(d[-1])
            list_data.append(d)
```

```python
list_data
```

```python
list_tag
```

```python
stopwords=[]
with open("stopwords.txt") as stops:
    for i in stops:
        print(i.strip())
        stopwords.append(i.strip())
```

```python
stopwords
```

```python
listdata=[]
for i in list_data:
    li=[]
    for j in i:
        if j not in stopwords:
            #print(j)
            li.append(j)
    listdata.append(li)
```

```python
listdata
```

```python
from collections import defaultdict
freq=defaultdict(int)
for i in listdata:
    for j in i:
        freq[j]=freq[j]+1
freq
```

```python
refined_data=[]
for i in listdata:
    temp=[]
    for j in i:
        #print(freq[j])
        if freq[j]>1:
            temp.append(j)
    refined_data.append(temp)
print(refined_data)
```

```python
count=0
for i in refined_data:
    if len(i)==0:
        refined_data.remove(i)
        list_tag.pop(count)
    count=count+1
```

```python
refined_data
```

```python
list_tag
```

```python
tags=[]
for i in list_tag:
    if i=='comment':
        tags.append(0)
    elif i=='complaint':
        tags.append(1)
    elif i=='meaningless':
        tags.append(2)
    elif i=='request':
        tags.append(3)
    elif i=='bug':
        tags.append(4)
    else:
        tags.append(5)
```

```python
tags
```

```python
import sklearn
from sklearn.model_selection import KFold
```

```python
kf=KFold(n_splits=10,shuffle=True)
```

```python
kf
```

```python
dataset=dense_tfidf
```

```python
sd=list(kf.split(dataset))
```

```python
sd
```

```python
from sklearn import svm
#from sklearn.naive_bayes import GaussianNB
#from sklearn.naive_bayes import BernoulliNB
#from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
#from sklearn import tree
#from sklearn.ensemble import RandomForestClassifier
f1=[]
presi=[]
recall=[]
acc_list=[]
for train,test in sd:
    #print(train,test)
    #dividing training and testing data using indexes we generated above
    train_data=[]
    train_label=[]
    test_data=[]
    test_label=[]
    for i in train:
        train_data.append(dataset[i])
        train_label.append(tags[i])
    for i in test:
        test_data.append(dataset[i])
        test_label.append(tags[i])
    #print(train_data)
    #print(test_label)
    classifier=svm.SVC() #support vector classifier,, we r actually plotting the graph but in mathematcial form
    #classifier=GaussianNB()
    #classifier=BernoulliNB()
    #classifier=MultinomialNB()
    #classifier=KNeighborsClassifier(n_neighbors=12)
    #classifier=tree.DecisionTreeClassifier()
    #classifier=RandomForestClassifier(n_estimators=100,max_depth=2,random_state=0)
    classifier.fit(train_data,train_label)#we r training our model
    predicted_labels=classifier.predict(test_data)
    incorrect=(predicted_labels!=test_label).sum()
    accuracy=(len(test_label)-incorrect)/len(test_label)
    #print(type(test_label),type(predicted_labels))
    score = f1_score(test_label,predicted_labels,average='weighted')
    f1.append(score)
    precision=precision_score(test_label,predicted_labels, average='weighted')
    presi.append(precision)
    rec=recall_score(test_label,predicted_labels, average='weighted')
    recall.append(rec)
    print(score,'\t\t',precision,'\t\t',rec)
    acc_list.append(accuracy)
maxi=max(acc_list)
print('\n\n',maxi)
```
